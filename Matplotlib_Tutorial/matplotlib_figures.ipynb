### Matplotlib - Object-oriented Interface / Figures  and axis

The main idea behind using the more formal object-oriented method is to create figure objects and then just call methods or attributes off of that object. This approach helps better in dealing with a canvas that has multiple plots on it.

```python
fig = plt.figure()
```

### Fig Size

```python
plt.figure(figsize=(14,5))
```

Now add axes to figure. The add_axes() method requires a list object of 4 elements corresponding to left, bottom, width and height of the figure. Each number must be between 0 and 1 âˆ’

```python
ax=fig.add_axes([0,0,1,1])
```

```python
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
plt.show()
```
